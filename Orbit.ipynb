# Integration of earth orbit angular speed

```python
def ode(th,t):
    """ Ode that describes the time evolution of the angle of earths posistion along
    the orbit"""
    th0 = 0
    G = 6.67384e-11
    M = 1.9891e30
    m = 5.9736e24
    e = 0.01671123
    a = 149598261e3
    h2 = a*(1+e**2)*G*(M+m)
    r = h2/(G*M)/(1+e*cos(th-th0))
    return sqrt(h2)/r**2
```

```python
from scipy.integrate import odeint
t = linspace(0,1.000017421*365.25*24*3600,1000)
thint = 0
th = odeint(ode,thint,t)

figure(figsize=(15,10))
beta = th[-1]/(1.000017421*365.25*24*3600)
plot(t/24/3600,(th.flatten()-beta*t)*180/pi)
ylabel('anglular position anomaly (deg)')
xlabel('clock time (d)')
```

```python
figure(figsize=(15,10))
plot(t/24/3600,(th.flatten()-beta*t)/2/pi*24*60)
print max((th.flatten()-beta*t)/2/pi*24*60)
```
