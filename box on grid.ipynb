# Finding grid points in an inclined box

Trying to find a fast algorithm. Let's start by drawing a picture

```python
x = range(-10,10,1)
y = range(-2,9,1)
X,Y = meshgrid(x,y)
plot(X,Y,'r.')
plot(0,0,'bo');

# defining a box
box_width = 2.4
box_length = 7.4
box_inclination = -30./180*pi
box_upright = array([[-box_width/2.,0],\
                     [box_width/2.,0], \
                     [box_width/2.,box_length],\
                     [-box_width/2.,box_length]])
rotation_matrix = array([[cos(box_inclination),-sin(box_inclination)],\
                         [sin(box_inclination),cos(box_inclination)]])
box_inclined = transpose(dot(rotation_matrix,transpose(box_upright)))
polygon = Polygon(box_inclined, facecolor='k', alpha=0.5)
gca().add_artist(polygon)
axis('equal')
```

The problem is to find the grid points inside the box in an efficient way. Here is one way, using matplotlibs 
fast points_in_poly function

```python
from matplotlib.nxutils import points_inside_poly

def points_in_box(X,Y,box):
    xx = X.flatten()
    yy = Y.flatten()
    xypoints = [(xx[k],yy[k]) for k in range(len(xx))]
    xyverts = [(box[k,0],box[k,1]) for k in range(4)]
    inside = points_inside_poly(xypoints,xyverts)
    xyin = array(xypoints)[inside]
    inside.shape = X.shape
    return xyin,inside
```

This is how it does for the example above

```python
x = range(-10,10,1)
y = range(-2,9,1)
X,Y = meshgrid(x,y)
plot(X,Y,'r.')
plot(0,0,'bo');

# defining a box
box_width = 2.4
box_length = 7.4
box_inclination = -30./180*pi
box_upright = array([[-box_width/2.,0],\
                     [box_width/2.,0], \
                     [box_width/2.,box_length],\
                     [-box_width/2.,box_length]])
rotation_matrix = array([[cos(box_inclination),-sin(box_inclination)],\
                         [sin(box_inclination),cos(box_inclination)]])
box_inclined = transpose(dot(rotation_matrix,transpose(box_upright)))
polygon = Polygon(box_inclined, facecolor='k', alpha=0.5)
gca().add_artist(polygon)
axis('equal')

xyin,inside = points_in_box(X,Y,box_inclined)
scatter(xyin[:,0],xyin[:,1])
```
