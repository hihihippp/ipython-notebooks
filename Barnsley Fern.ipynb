### Barnsley Fern

The Barnsley Fern is a fractal named after the British mathematician Michael Barnsley. 
He made it to resemble the Black Spleenwort, Asplenium adiantumnigrum.

It is a nice fractal based on randomly choosing between 4 affine transformations. It serves to
illustrate pythons capability

```python
from numpy.random import rand
N = 5000000
fernM1 = array([[0,0],[0,0.16]])
fernM2 = array([[0.2,-0.26],[0.23,0.22]])
fernM3 = array([[-0.15,0.28],[0.26,0.24]])
fernM4 = array([[0.85,0.04],[-0.04,0.85]])
fern = [fernM1,fernM2,fernM3,fernM4] 
dd = rand(N)
allZ = zeros((2,N))

def bin(a):
    bins = [0.025,0.125,0.225]
    x = 0
    for b in bins:
        if a > b:
            x += 1
    return x

where = map(bin,dd)
a = array([[0,0,0,0],[0,1.6,0.44,1.6]])
z = array([0,1])
for k in range(N):
    w = where[k]
    z = dot(fern[w],z).T+a[:,w]
    allZ[:,k] = z
```

```python
figsize(15,12)
plot(allZ[0,:],allZ[1,:],'.g',ms=0.1);
```
