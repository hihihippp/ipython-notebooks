## Determine the center of an interpolating circle given 3 points.

Given three points on the perimeter of a circle the parameters of the circle (radius and center) are uniqely determined. 

<img src="http://upload.wikimedia.org/wikipedia/commons/8/88/Triangle_%28Acute%29_Circumscribed.svg">

This is a simple function to do this that I wrote for my spring mesher.

```python
from pylab import *
import warnings

def circle_center(p0,p1,p2):
    
    A = array([p0,p1,p2])

    # Handle the possibility that the plane defined by the three points goes through origo. 
    if cond(A) > 1e10:
        offset = cross(p1-p0,p2-p0)
    else:
        offset = array([0.,0.,0.])

    px0=p0[0]+offset[0]
    py0=p0[1]+offset[1]
    pz0=p0[2]+offset[2]
    px1=p1[0]+offset[0]
    py1=p1[1]+offset[1]
    pz1=p1[2]+offset[2]
    px2=p2[0]+offset[0]
    py2=p2[1]+offset[1]
    pz2=p2[2]+offset[2]

    A = array([[px0,py0,pz0],
               [px1,py1,pz1],
               [px2,py2,pz2]])

    d = 100.
    a = array([d,d,d])
    
    if cond(A) == inf:
        raise LinAlgError("Points are colinear")
        
    if cond(A) > 1e10:
        warnings.warn("The points are close to colinear")
        
    abc = solve(A,a)  # The coefficients of the plane

    B = array([ [-px0/2. + px1/2., -py0/2. + py1/2., -pz0/2. + pz1/2.],
                [-px1/2. + px2/2., -py1/2. + py2/2., -pz1/2. + pz2/2.],
                list(abc) ])
    b = array([ -(px0**2/4. - px1**2/4. + py0**2/4. - py1**2/4. + pz0**2/4. - pz1**2/4.),
                -(px1**2/4. - px2**2/4. + py1**2/4. - py2**2/4. + pz1**2/4. - pz2**2/4.),
                                                                                d])

    x = solve(B,b) - offset
    return x[0],x[1],x[2]
```

### Test the function for some random points

```python
for i in range(3):
    p0 = rand(3)
    p1 = rand(3)
    p2 = rand(3)

    cx,cy,cz = circle_center(p0,p1,p2)
    c = array([cx,cy,cz])

    # derive a orthonormal base
    u0 = (p0-c)/norm(p0-c)
    u1 = p1-c - dot(p1-c,u0)*u0
    u1 = u1/norm(u1)
    
    # plot the circle, the center and the points
    R = norm(p0-c)
    xp = R*cos(linspace(0,2*pi,100))
    yp = R*sin(linspace(0,2*pi,100))
    xyz = ones((100,3))*c + xp[:,np.newaxis]*u0 + yp[:,np.newaxis]*u1
    plot(xyz[:,0],xyz[:,1])
    plot(c[0],c[1],'o')
    plot([p0[0],p1[0],p2[0]],[p0[1],p1[1],p2[1]],'s')
```
