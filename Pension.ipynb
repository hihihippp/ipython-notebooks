So we want to be on the efficient frontier. For convenience we pick this, which is based on historical data for a mix of S&P 500 index 
fonds and US treasury 10 yar bonds (I only estimated the coefficients based on a similar diagram I saw in Shillers Open Yale Finance 
course. But it is roughly right).

```python
a = .014
b = .06
s = linspace(0,0.20,1000)
r = a*sqrt(s**2/b**2-1)+0.1
plot(s,r,s,a/b*s+0.1,'k--')
axis("equal")

def interest_efficient_frontier(risk):
    return a*sqrt(risk**2/b**2-1)+0.1
```

So what happens if we average a couple of thousand lives will we get something similar to a life with zero volatility.

```python
risk = lambda y: 8./100
interest_rate = lambda y: interest_efficient_frontier(risk(0))
age_start = 30
salary_start = 33000.*12
age_end = 65
salary_end = 60000.*12
saving_fraction = 4000.0/33000

length_career = age_end-age_start
raise_rate = log(salary_end/salary_start)/length_career  # salary increase rate (assuming exponential growth)

t,true_avg_cum_savings = life(interest_rate, lambda y: 0.0, age_start, age_end, salary_start, salary_end)

avg_cum_savings = zeros(len(true_avg_cum_savings))
cum_variance = zeros(len(true_avg_cum_savings))

for k in range(20000):
    t,cum_savings = life(interest_rate, risk, age_start, age_end, salary_start, salary_end)
    avg_cum_savings = avg_cum_savings*float(k)/(k+1) + cum_savings/(k+1)
    cum_variance = cum_variance*float(k)/(k+1) + ((cum_savings-true_avg_cum_savings)**2)/(k+1)

cum_std = sqrt(cum_variance)

plot(t,avg_cum_savings*1e-6,label="average savings")
plot(t,cum_savings*1e-6,label="a random life")
plot(t,(avg_cum_savings-cum_std)*1e-6,'k--',label="1 std dev")
legend(loc=2)
plot(t,(avg_cum_savings+cum_std)*1e-6,'k--',label="1 std dev")
xlabel("saving time (years)")
ylabel("Savings (million SEK)")
print("low estimate end savings: %0.2f Mkr" % ((avg_cum_savings[-1]-cum_std[-1])/1e6))
```
