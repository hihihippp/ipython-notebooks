# Ideal price of stock for a known future dividend

Suppose the future divident of an general index fund is a known function of time. How would that affect the price of the share.

The idea here is to apply the theory of discounted cash flow to some common continous functions to see what happens.

Suppose the rate of divident is $d(t)$. Then the theory of discounted cash flow suggests the the price now should be
$$ p = \int_0^\infty d(t) e^{-\lambda t} dt $$
where $\lambda = \log(1+r)$.

```python
r = 4.0/100 # Safe interest rate
lam = log(1+r)

from scipy.integrate import quad
```

Suppose $d(t) = \frac{1}{2}(1+\sin t)$. What happens?

```python
def d(t):
    return 0.5*(1+sin(t))

def plot_price_and_dividend(d):
    t = linspace(0,10*pi,1000)
    p = zeros(1000)
    for ii in range(len(t)):
        p[ii] = quad(lambda tau: d(t[ii]+tau)*exp(-lam*tau),0,500,limit=1000)[0]
    plot(t,p,'b', t,d(t),'g')
    
plot_price_and_dividend(d)
```

Interesting! In markets where dividents behave this way the price seem be $\pi/4$ before the dividend.

Now, suppose we have a market where it either is a particular dividend or it ain't. 
$$ 
d(t)= \cases{
1  & \text{ if } \sin(t)\ge 0 \cr
0  & \text{ if } \sin(t)\lt 0
} $$

```python
from scipy.special import erf
def d(t):
    return 0.5*(1+erf(5*sin(t)))

plot_price_and_dividend(d)
```

Aha! From the rectangular wave in the dividend rate we get something like a triangular wave in the price. The high point of the price
seem to coincide with beginning of a period of nice dividends  and the low point in price with the beginning of a bad period. Quite 
natural. It also seems the price is about 30 years worth of dividends.

It surprises me that the price simply looks like an offset version of $\int d(t) dt$ alone. I.e., the price looks like 
$p(t)\approx b + \int_0^t d(\tau)d\tau$. To understand this we plot the integrand

```python
t = linspace(0,300,1000)
plot(t,d(t)*exp(-lam*t))
```

Aha! We see that for this frequently alternating $d$ the exponential factor dominates the shape of the integrand.

I wonder what happens if the period of $d$ is five times longer

```python
def d(t):
    """ exaxt function. May be difficult to integrate """
    return 0.5*(1+erf(50*sin(0.2*t)))

plot_price_and_dividend(d)
```

How about if the rate of dividend increases linearly. Will the price then appear to increase quadratically. That would be weird. 
Wouldn't it?

```python
def d(t):
    return t

plot_price_and_dividend(d)
```

Apparently a linear increase in rate of dividend results in a linear increase in price. Logical but it nuances the idea that the
price is the integral of the dividends. That is apparently not true for increasing dividends

## S & P data

Now lets see how the stockmarket has done trying to estimate future dividends. I got the S&P 500 data from Robert Shillers homepage

```python
def fix(x):
    if x == "NA":
        return NaN
    else:
        return x
crs = {}
for c in range(11):
    crs[c] = fix

data = loadtxt("ie_data.csv",converters=crs,skiprows=9)
```

```python
tt = data[:,5]
pp = data[:,7]
dd = data[:,8]
plot(tt,pp,tt,dd)
xlabel("time (years)")
ylabel("Price index")
```

We plot the historical dividend as well. For sake of computing DCF we also make a trend line that we assume will hold
for for future dividends

```python
tfut = linspace(tt[-1]+0.1,2100,100)
dfut = 0.15*(tfut-2012)+25

plot(tt,dd,'g',tfut,dfut,'g:')
xlabel("time (years)")
ylabel("price index")
```

```python
r =6.0/100 # Safe interest rate
lam = log(1+r)

tt_ext = concatenate((tt,tfut))
dd_ext = concatenate((dd,dfut))

from scipy.integrate.quadrature import cumtrapz
t = zeros(len(tt))
p = zeros(len(tt))
for ii in range(len(pp)):
    t[ii] = tt[ii]
    p[ii] = cumtrapz(dd_ext[ii:]*exp(-lam*(tt_ext[ii:]-tt_ext[ii])),x=tt_ext[ii:])[-1]
plot(tt,pp,'b', t,p,'r')
xlabel('time (years)')
ylabel('price index')
legend(("S&P 500 index","DCF of dividends"))
```

The model works well for 1870-1980 if a risk free interest rate of 6 % is assumed.  Lets see what the risk free interest should be
for the model to fit the data for todays index:

```python
r =2.0/100 # Safe interest rate
lam = log(1+r)

tt_ext = concatenate((tt,tfut))
dd_ext = concatenate((dd,dfut))

from scipy.integrate.quadrature import cumtrapz
t = zeros(len(tt))
p = zeros(len(tt))
for ii in range(len(pp)):
    t[ii] = tt[ii]
    p[ii] = cumtrapz(dd_ext[ii:]*exp(-lam*(tt_ext[ii:]-tt_ext[ii])),x=tt_ext[ii:])[-1]
plot(tt,pp,'b', t,p,'r')
```

Answer: about 2 %. Weird! Or is it that the stock market is overvalued? Or is it that the dividends will increase even faster in the

Why not making a better approach for the influence of a time dependant risk free interest rate. I guess the DCF for a time interval
$dt$ will be
$$ dp = d(t) \exp \left( -\int_0^t \lambda(\tau) d\tau \right) dt $$ 
where $\lambda(t) = \log(1+r(t))$ and $r$ is the risk free interest rate. So the price will be
$$ p = \int_0^\infty d(t) \exp \left( -\int_0^t \lambda(\tau) d\tau \right) dt $$

```python
def dcf(d,r,maturity=Inf):
    """ Computes the discounted cash flow price of a share giving continous 
        dividend d subject to safe interest rate r, which may be a specified
        vectorized function of time or be a constant. """
    try:
        constant_r = False
        r(t)
    except:
        constant_r = True
    
    if constant_r:
        lam = log(1+r)
        return quad(lambda t: d(t)*exp(-lam*t),0,maturity,limit=1000)[0]
    else:
        lam = lambda t: log(1+r(t))
        int_lamdt = vectorize(lambda t: quad(lam, 0,t,limit=1000)[0])
        return quad(lambda t: d(t)*exp(-int_lamdt(t)),0,maturity,limit=1000)[0]
        
    
        
        
```

```python
def d(t):
    return 0.5*(1+sin(t))

r = lambda t: 3.0/100
    
import sys

def plot_price_and_dividend(t,d,r):
    p = zeros(len(t))
    for ii in range(len(t)):
        if ii % 100 == 0:
            sys.stdout.write("[%d]" % (ii/10.0))
        p[ii] = dcf(lambda tao: d(tao+t[ii]),lambda tao: r(tao+t[ii]))
    plot(t,p,'b', t,d(t),'g')

t = linspace(0,10*pi,1000)
plot_price_and_dividend(t,d,r)
```

```python
def d(t):
    return 0.15*t+5

def r(t):
    rmin = 2.0/100
    rmax = 6.0/100
    tcrit = 140
    return rmin + (rmax-rmin)*0.5*(1-erf(0.1*(t-tcrit)))

plot(t+70,r(t+70))
```
