# Introduktion till Statistiska centralbyråns nya API

Under årets Almedalsvecka blev jag medveten om att Statistiska centralbyrån har släppt ett API för att komma åt statistikdatabasen. Länk finns <a href="http://www.scb.se/api" target="tab">här</a>. APIet är baserat på JSON och är tillräkligt generellt för att kunna söka med wildcards etc. 

Detta är ju både trevligt och användbart om man kan slippa rota i excel dokument som tidigare varit min erfarenhet med SCB. Min tanke här är att se om jag kan få detta att fungera. 


## Ett snabbt första exempel

Som ett första exempel tänkte jag att vi kunde studera hur befolkningsutvecklingen varit på Gotland samt hur det kan jämföras med befolkningsutvecklingen i Riket samt i Kalmars län.

```python
import urllib2
import json

base_url = "http://api.scb.se/OV0104/v1/doris/sv/ssd"
table = "/BE/BE0101/BE0101A/BefolkningNy"
query = """
{  
    "query": [
        {       
            "code": "ContentsCode",
            "selection": {        
                "filter": "item",         
                "values": [          
                  "BE0101N1"         
                ]      
            }    
        },    
        {      
            "code": "Region",
            "selection": {        
                "filter": "item",         
                "values": [
                    "00",
                    "08",
                    "09"
                ]      
            }    
        }   
    ],  
    "response": {    
        "format": "json",  
    }
}"""

http_request = urllib2.Request(url=base_url+table, data=query)
http_result = urllib2.urlopen(http_request).read()
bef_all = json.loads(http_result.decode('utf-8-sig'))
```

Med hjälp av **numpy** och **matplotlib** så kan vi enkelt plotta befolkningsutvecklingen i de olika länen samt i riket. 

```python
from pylab import *

# Filtrera ut de data vi är intresserade av och gör om till numpy arrayer
year = array(map(lambda x:int(x['key'][1]),filter(lambda x:x['key'][0]==u"Gotlands län", bef_all['data'])))
bef_got = array(map(lambda x:float(x['values'][0]),filter(lambda x:x['key'][0]==u"Gotlands län", bef_all['data'])))
bef_kal = array(map(lambda x:float(x['values'][0]),filter(lambda x:x['key'][0]==u"Kalmar län", bef_all['data'])))
bef_rik = array(map(lambda x:float(x['values'][0]),filter(lambda x:x['key'][0]==u"Riket", bef_all['data'])))

# Plotta
figure(figsize=(13,7))
plot(year,bef_got/bef_got[0],label=u'Gotlands län')
plot(year,bef_kal/bef_kal[0],label=u'Kalmar län')
plot(year,bef_rik/bef_rik[0],label=u'Riket')
legend(loc='best')
title(u'Befolkning normaliserat till 1968 års befolkning')
xlabel(u"År")
ylabel(u"Befolkning (normaliserad med 1968 års nivåer)");
```

## Hjälpfunktioner för att ställa frågor till databasen och presentera resultaten

För att slippa repetera kod definieras först några hjälpfunktioner. Syftet är att förenkla förfarandet med att skicka förfrågningar och skriva ut resultaten.

```python
from IPython.display import HTML
import pprint

def scb_html_table(res):
    """Skriver ut resultat från SCB-frågor till HTML som renderas i IPython notebook"""
    try:
        # Let's assume we have a table
        tab = "<table>\n\t<tr>\n"
        for c in res['columns']:
            tab += "\t\t<th>" + c['text'] + "</th>\n"
        for r in res['data']:
            tab += "\t</tr>\n\t<tr>\n"
            for k in r['key']:
                tab += "\t\t<td>" + k + "</td>"
            for k in r['values']:
                tab += "<td>" + k + "</td>\n" 
            tab += "\t</tr>\n"
    except:
        try:
            # then we probably have a list of similar dicts
            tab = "<table>\n\t<tr>\n"
            for c in res[0].keys():
                tab += "\t\t<th>" + c + "</th>\n"
            for r in res:
                tab += "\t</tr>\n\t<tr>\n"
                for c in r:
                    tab += "\t\t<td>" + r[c] + "</td>\n"
                tab += "\t</tr>\n"   
        except:
            # then I guess we are looking at the metadata of a table
            tab += '<h3>' + res['title'] + "</h3>\n"
            tab += "<h4>Variables:</h4>\n"
            for v in res['variables']:
                tab += "<table>"
                tab += "<tr><th>code</th><td>"+v['code']+"</td></tr>\n"
                tab += "<tr><th>text</th><td>"+v['text']+"</td></tr>\n"
                tab += "<tr><th>valueText</th><th>value</th></tr>\n"
                for region,kod in zip(v['valueTexts'],v['values']):
                    tab += "<tr><td>"+region+"</td><td>"+kod+"</td></tr>\n"   
                tab += "</table>\n"
    return HTML(tab)

import urllib2
import json

def scb_query(table="",query=""):
    """Ställer frågor till statistikdatabasen"""
    base = "http://api.scb.se/OV0104/v1/doris/sv/ssd"
    req = urllib2.Request(url=base+table, data=query)
    f = urllib2.urlopen(req)
    if f.code == 200:
        print "Request succeeded"
        res = f.read().decode('utf-8-sig')
        res = json.loads(res)
        return res 
    else:
        raise ValueError("HTTP förfrågan returnerade kod: %d" % (f.code))
```

## Att hitta rätt tabell i Statistikdatabasen

Statistikdatabasen är hierarkiskt uppbyggd. Den översta nivån indelar tabellerna i breda kategorier. Man kan lätt lista statistikdatabasens ämnesområden genom:

```python
scb_html_table(scb_query())
```

För att illustrera processen med att hitta vad man söker tänkte jag börja med att försöka plocka ut statistik för ungdomsarbetslösheten. Detta med anledning av att så många politiker pratat om den under Almedalsveckan.

För att lista de subkategorier som ingår i ämnesområdet arbetsmarknad gör man på följande sätt: 

```python
scb_html_table(scb_query(table='/AM'))
```

Vi kan gå vidare och undersöka vilka tabeller som finns under **Arbetskraftsundersökningarna (AKU)**. Detta lär vara var vi kan hitta uppgifter om ungdomsarbetslöshet:

Vidare är vi intresserade av att jämföra ungdomsarbetslösheten så vi plockar ut tabelen om **Arbetslösa**

```python
scb_html_table(scb_query(table='/AM/AM0401/AM0401L'))
```

```python
res = scb_query(table='/AM/AM0401/AM0401L/NAKUArblosaTAr')
scb_html_table(res)
```

## Ställa frågor till databasen

För att få ut data ur databasen måste man avgöra vilken tabell som innehåller den behövda informationen samt konstrura en fråga som plockar ut rätt data ur tabellen. 

En fråga är en enkel datastruktur i JSON-format. 

```python
table='/AM/AM0401/AM0401L/NAKUArblosaTAr'
query = """
{  
    "query": [
        {       
            "code": "ContentsCode",
            "selection": {        
                "filter": "item",         
                "values": [          
                    "AM0401G0"         
                ]      
            }    
        },
        {       
            "code": "Alder",
            "selection": {        
                "filter": "item",         
                "values": [          
                     "15-24",
                     "tot15-74"
                ]      
            }    
        },

    ],  
    "response": {    
        "format": "json",  
    }
}"""
```

Med hjälp av hjälpfunktionen **scb_query** som jag definerade ovan så kan frågan enkelt ställas till databasen: 

```python
arb = scb_query(table=table,query=query)
```

Med lite trickande kan vi plocka ut ungdomsarbetslösheten samt jämföra den med den totala arbetslösheten.


```python
from pylab import *
y = array(map(lambda x:int(x['key'][1]),filter(lambda x:x['key'][0]==u"15-24 år", arb['data'])))
ung = array(map(lambda x:float(x['values'][0]),filter(lambda x:x['key'][0]==u"15-24 år", arb['data'])))
alla = array(map(lambda x:float(x['values'][0]),filter(lambda x:x['key'][0]!=u"15-24 år", arb['data'])))

width = 0.7
bar(y,ung,width,color='r',label=u'Unga')
bar(y,alla-ung,width,bottom=ung,label=u'Övriga')
xticks(y+width/2., map(str,y))
ylabel(u'Tusentals arbetslösa')
legend(loc='best')

scb_html_table(arb)
```
