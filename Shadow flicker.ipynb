# The Mathematics of shadow flicker

When I was working for a small wind turbine manufacturer I was asked to calculate the number of hours per year the
houses in the neighborhood of our prototype wind farm would be in the shadow of our turbine. Now, there are software that
calculates this sort of thing with a few clicks of the mouse. However, being mathematically inclined I instead solved the problem 
myself. 

## Outline of the calculations
Okay, so the general idea is this. 

1. Take a point that is located above the earth surface at a fixed longitude and latitude.
2. Given the longitude, latitude, time of year and the time of day find out where
this point will be in a universal coordinate system
3. In the universal coordinates, find where the shadow of this point will hit 
the sphere.
4. Transform this point on the sphere back to a longitude and latitude.

## The shadow of point

Let's say the earth surface is a sphere of radius $R$ centered at origo. Now we have a point above the surface which casts a shadow on the sphere from a lightsource at $y=+\infty$. Can we calculate where this shadow
will end up on the sphere. I'm no expert on geometry, but following wikipedia [Line sphere intersection](http://en.wikipedia.org/wiki/Line%E2%80%93sphere_intersection) it
seems pretty straight forward.

```python
def point_shadow(p):
    """ Returns the shadow of a point on a sphere centered in (0,0,0). 
    global variable R is the radius of the sphere an p is the cartesian
    coordinates of the point. """
    global R
    c = -array(p).flatten()
    v = array([0,-1,0])  
    d = dot(v,c) - sqrt(dot(v,c)**2-dot(c,c)+R**2)
    if d < 0 or isnan(d):
        return None
    else:
        return p+d*v
    
    
def point_shadow_plane(p):
    """ Returns the shadow of a point p cast onto a plane tangent to a sphere 
    centered in (0,0,0) and normal to the vector p. Global variable R is the 
    radius of the sphere and p is the cartesian coordinates of the point. """
    global R
    c = -array(p).flatten()
    n = -c/norm(c)
    p0 = R*n
    v = array([0,-1,0])  
    d = dot(p0-p,n)/dot(v,n)
    if d < 0 or isnan(d):
        return None
    else:
        return p+d*v
```

```python
# Check that we are really on the surface of the sphere
R = 6371e3
p = R*array([0.7,0.7,0.7])
s = point_shadow(p)
# I.e., the following should be zero
R-sum(s**2)**0.5
```

That seems to work. 

### Coordinate transformations

We want a coordinate transformation from a coordinate rotating with earth 
(longitude $\phi$, latitude $\delta$) to a system fixed nonrotating system.

A point expressed in geographic coordinates is characterized by its longitude, 
latitude and height ($h$) over the surface. In the cartesian system rotating with 
earth this becomes

$ x = (R+h) \sin(\pi/2-\delta)\cos \phi $

$ y = (R+h) \sin(\pi/2-\delta)\sin \phi $

$ z = (R+h) \cos (\pi/2 - \delta) $

What we want to do next is to transform a point expressed this way into the 
nonrotating system.


#### Transformation to nonrotating system and back

The transformation matrix for rotation about the z-axis is

$$
R_z(\Omega t) = \left\[ \begin{array}{ccc}
\cos \Omega t & -\sin \Omega t & 0 \\\\
\sin \Omega t & \cos \Omega t & 0 \\\\
0 & 0 & 1
\end{array} \right\] 
$$

where $\Omega$ is the rotational speed of earth $=2 \pi / 86164.098903691$ rad/s.  

Next we want to express this in a system aliged with the earth orbital plane. 
Thus we want to rotate about the universal y-axis by the axial tilt angle 
$\alpha$ ($= 23.4 {}^\circ$). The transformation matrix for this rotation is

$$
R_x(\alpha) = \left\[ \begin{array}{ccc}
1 & 0 & 0 \\\\
0 & \cos \alpha t & -\sin \alpha t \\\\
0 & \sin \alpha t & \cos \alpha t
\end{array} \right\] 
$$

Finally we want to rotate this system so we face a around the sun.

$$
R_z(\beta t) = \left\[ \begin{array}{ccc}
\cos \beta t & -\sin \beta t & 0 \\\\
\sin \beta t & \cos \beta t & 0 \\\\
0 & 0 & 1
\end{array} \right\] 
$$
where $\beta$ is the earth orbital speed ($=2 \pi/(1.000017421 \cdot 365.25 \cdot 24 \cdot 3600)$ rad/s)

The total transformation from ($\phi$,$\delta$) is thus

$$
\left\[\begin{array}{c}
X \\\\
Y \\\\
Z
\end{array}\right\] = T(\phi,\delta) = R_z(-\beta t) R_x(\alpha) R_z(\Omega t) \left\[\begin{array}{c}
(R+h) \sin(\pi/2-\delta)\cos \phi \\\\
(R+h) \sin(\pi/2-\delta)\sin \phi \\\\
(R+h) \cos (\pi/2 - \delta)
\end{array}\right\]
$$

The inverse of this tranformation is obtained by first inversing the rotations
$$
\left\[\begin{array}{c}
x \\\\
y \\\\
z
\end{array}\right\] = \left(R_z(\beta t) R_x(\alpha) R_z(\Omega t)\right)^{-1} \left\[\begin{array}{c}
X \\\\
Y \\\\
Z
\end{array}\right\]
$$
and then the transformation to geographic coordinates

$h=\sqrt{x^2+y^2+z^2}-R$ 

$\phi  = \mbox{arctan}\left(\frac{y}{x}\right)$

$\delta = \pi/2 - \mbox{arccos}\left(\frac{z}{r}\right)$

Let's see if it works.

```python
def Rx(t):
    return matrix([[1,0,0],[0,cos(t),-sin(t)],[0,sin(t),cos(t)]])

def Ry(t):
    return matrix([[cos(t),0,sin(t)],[0,1,0],[-sin(t),0,cos(t)]])

def Rz(t):
    return matrix([[cos(t),-sin(t),0],[sin(t),cos(t),0],[0,0,1]])


R = 6371e3
alpha = 23.439/180*pi
beta = 2*pi/(1.000017421*365.25*24*3600)  # source wikipedia
Omega = 2*pi/86164.098903691

def Geographical_to_Earth_Cartesian(phi,delta,h):
    global R
    return matrix([[(R+h)*sin(pi/2-delta)*cos(phi)],\
                   [(R+h)*sin(pi/2-delta)*sin(phi)],\
                   [(R+h)*cos(pi/2-delta)]])


def Earth_Cartesian_to_Geographical(xyz):
    global R
    r = norm(xyz)
    x = xyz[0,0]
    y = xyz[1,0]
    z = xyz[2,0]
    return arctan2(y,x),pi/2-arccos(z/r),r-R
        

def Geographical_to_Universal(phi,delta,h,t):
    global R
    global alpha
    global beta
    global Omega
    return Rz(-beta*(t+160000*sin(beta*t)))*Rx(alpha)*Rz(Omega*t-pi/2)*Geographical_to_Earth_Cartesian(phi,delta,h)

def Universal_to_Geographical(XYZ,t):
    global R
    global alpha
    global beta
    global Omega
    xyz = inv(Rz(-beta*(t+160000*sin(beta*t)))*Rx(alpha)*Rz(Omega*t-pi/2))*XYZ
    return Earth_Cartesian_to_Geographical(xyz)  
```

## Example 1

Lets study the path the shadow of a point trace for the first day of the first six months of the year

```python
phi = 15./180*pi
delta = 60./180*pi
h = 1.

import sys

def plot_day_in_month_path(n):
    global phi
    global Omega
    global delta
    global h
    month_names = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
    t0 = n*30*2*pi/(Omega+beta)
    t = arange(t0,t0+24*3600,600)
    sphis = zeros(len(t))
    sdeltas = zeros(len(t))

    for k in range(len(t)):
        if (k % (24*60)) == 0:
            sys.stdout.write('.')
        pXYZ = Geographical_to_Universal(phi,delta,h,t[k])
        sXYZ = transpose(matrix(point_shadow(array(pXYZ).flatten())))
        if sXYZ[0,0] != None:
            sphi,sdelta,sh = Universal_to_Geographical(sXYZ,t[k])
            sphis[k] = sphi
            sdeltas[k] = sdelta
        else:
            sphis[k] = None
            sdeltas[k] = None
    plot((sphis-phi)*R*cos(delta),(sdeltas-delta)*R,'o-',label="1st of "+month_names[n])
    
for n in range(7):
    plot_day_in_month_path(n)
            
plot(0,0,'ks')
axis('equal')
xlabel('East-west (m)')
ylabel('South-north (m)')
legend()

xlim((-4,4))
ylim((-10,16));
savefig("shadowtrack.png",dpi=150)
```

## Example: an Analemma

```python
phi = 0./180*pi
delta = 51.4791/180*pi
h = 1.

import sys

def analemma(hourZ):
    t0 = hourZ*3600
    t = arange(t0,t0+365*24*3600,24*3600)
    sphis = zeros(len(t))
    sdeltas = zeros(len(t))

    for k in range(len(t)):
        if (k % (24*60)) == 0:
            sys.stdout.write('.')
        pXYZ = Geographical_to_Universal(phi,delta,h,t[k])
        sXYZ = transpose(matrix(point_shadow(array(pXYZ).flatten())))
        if sXYZ[0,0] != None:
            sphi,sdelta,sh = Universal_to_Geographical(sXYZ,t[k])
            sphis[k] = sphi
            sdeltas[k] = sdelta
        else:
            sphis[k] = None
            sdeltas[k] = None
            
    return (sphis-phi)*R*cos(delta),(sdeltas-delta)*R
        
plot(0,0,'ks')
for hour in [12]:
    x,y = analemma(hour)
    plot(x,y,'-')
    
axis('equal')
xlabel('East-west (m)')
ylabel('South-north (m)')
```

```python
for hour in [12]:
    x,y = analemma(hour)
    r = (x**2+y**2)**0.5
    azi = arctan2(x,y)
    alt = arctan2(h,r)
    plot(azi*180/pi,alt*180/pi,label=str(hour))
legend()
grid(True)
```

## Shadow of a solid object

Now we want to take into account that we want to find the shadow of a solid object. To simplify matters we
assume that the shadow of the object is a rectangle on the ground. The object is shown as a red dot and the tip
of the shadow is shown as a blue dot. The shadow is the grey rectangle.

```python
plot(0,0,'ro',label='Object')
plot(4,6,'bo',label='Shadow of top point')
def shadow_box(width,tip_xy):
    box_width = width
    box_length = norm(tip_xy)
    box_inclination = -arctan2(float(tip_xy[0]),tip_xy[1])
    box_upright = array([[-box_width/2.,0],\
                         [box_width/2.,0], \
                         [box_width/2.,box_length],\
                         [-box_width/2.,box_length]])
    rotation_matrix = array([[cos(box_inclination),-sin(box_inclination)],\
                             [sin(box_inclination),cos(box_inclination)]])
    box_inclined = transpose(dot(rotation_matrix,transpose(box_upright)))
    return box_inclined

box_inclined = shadow_box(1.2,(4,6))
polygon = Polygon(box_inclined, facecolor='k', alpha=0.5,label='Shadow')
gca().add_artist(polygon)
axis('equal')
xlim((-1,6))
ylim((-1,8))
```

```python
from matplotlib.nxutils import points_inside_poly

def points_in_box(X,Y,box):
    xx = X.flatten()
    yy = Y.flatten()
    xypoints = [(xx[k],yy[k]) for k in range(len(xx))]
    xyverts = [(box[k,0],box[k,1]) for k in range(4)]
    inside = points_inside_poly(xypoints,xyverts)
    xyin = array(xypoints)[inside]
    inside.shape = X.shape
    return xyin,inside
```

```python
phi = 15./180*pi
delta = 60./180*pi
h = 50.
width = 2.

x = linspace(-200,200,100)
y = linspace(-100,150,100)
X,Y = meshgrid(x,y)
```

```python
import sys
shadow_time = zeros(X.shape)
t = arange(0,365*24*3600,60)

for k in range(len(t)):
    if (k % (24*60)) == 0:
        sys.stdout.write("%3d " % (k/24/60+1))
        if (k % (25*24*60)) == 0:
            if k != 0:
                sys.stdout.write("\n")
    pXYZ = Geographical_to_Universal(phi,delta,h,t[k])
    sXYZ = transpose(matrix(point_shadow(array(pXYZ).flatten())))
    if sXYZ[0,0] != None:
        sphi,sdelta,sh = Universal_to_Geographical(sXYZ,t[k])
        box = shadow_box(width,((sphi-phi)*R,(sdelta-delta)*R))
        xyin,inside = points_in_box(X,Y,box)
        shadow_time[inside] += 60. 

CS = contourf(X, Y, shadow_time/3600.,levels=[0.1,5,10,15,30,40,50,100],cmap=cm.get_cmap(name="gist_heat_r"))

xlim((-200,200))
ylim((-100,150));

savetxt("shadowtime.txt", shadow_time)
```
