# Modelling combustion in a closed boc

This is a simple model I have constructed to better understand combustion as it occurs in a 
combustion engine. It doest really work yet, as it produces super high temperaturers and 
pressures very fast. 

The general idea is to use classical kinetic gas theory to estimate the number of collisions 
in a mixture of petrol (represented by $C_8H_18$) and Oxygen ($O_2$) as a function of
temperature, volume and mass. The number of collisions is then assumed to be proportional to 
the number of reactions that take place. 

The reaction is assumed to be

$$2 C_8H_18 + 25 O_2 \rightarrow 16 CO_2 + 18 H_2O$$

The code below should not be very hard to understand. If it is, it is probably wrong.

There is a posibility to vary the volume during the combustion. However, I have not done this 
yet because I find it unwise to do that before I am sure about the simpler part of the model.

```python
k = 1.3807e-23 # Boltzmanns constant
d = 3e-10 # average molecule diameter
m = 1.4978e-25 # average molecular weight
g = 1e-9 # ratio of collisions that result in a reaction
k1 = g*pi*d**2*sqrt(2*k/pi/m)
E = 46.8e6 # energy content (J/kg) of gasoline
k2 = 2*E*m/16 # energy released for every CO2 created
M = 1e-3 # mass of petrol to simulate (kg)
NA0 = M/m # initial number of C8H18
V = 50e-6 # cylinder volume

def dNdt(N,t=0):
    """ The differential equation governing the system """
    U = k2*N[2]
    T = 2./3.*U/k/sum(N)+293.0
    return array([-2*k1*sqrt(T)*N[0]*N[1]/V,
                  -25*k1*sqrt(T)*N[0]*N[1]/V,
                  16*k1*sqrt(T)*N[0]*N[1]/V,
                  18*k1*sqrt(T)*N[0]*N[1]/V])
```

```python
from scipy import integrate
t = linspace(0, 1e-2,  1000)
N0 = array([NA0,25.0/2.*NA0,0.,0.])
N, infodict = integrate.odeint(dNdt, N0, t, full_output=True)
infodict['message']
```

```python
NA, NB, NC, ND = N.T
plot(t, NA, 'r-', label='Gasolene')
plot(t, NB, 'b-', label='Oxygen')
plot(t, NC, 'g-', label='Carbon Dioxide')
plot(t, ND, 'y-', label='Water')
grid()
legend(loc='best')
xlabel('time')
ylabel('population')
title('Evolution of species');
```

```python
U = k2*NC
T = 2./3.*U/k/(NA+NB+NC+ND)+293.0
plot(t,T,label='Temperature')
xlabel('time (s)')
ylabel('temperature (K)');
```

```python
Ntot = NA+NB+NC+ND
p = Ntot/V*k*T
plot(t,p/100000.,label='Pressure')
xlabel('time (s)')
ylabel('pressure (Bar)');
```
