# Modelling combustion in a closed box

This is a simple model I have constructed to better understand combustion as it occurs in a 
combustion engine. It doest really work yet, as it produces super high temperaturers and 
pressures very fast. 

The general idea is to use classical kinetic gas theory to estimate the number of collisions 
in a mixture of petrol (represented by $C_8H_18$) and Oxygen ($O_2$) as a function of
temperature, volume and mass. The number of collisions is then assumed to be proportional to 
the number of reactions that take place. 

The reaction is assumed to be

$$2 C_8H_{18} + 25 O_2 \rightarrow 16 CO_2 + 18 H_2O$$



## Derivation of the differential equations

The theory will be used here is called collison theory. We shall use the following symbols for the reaction speci:

* The number of $C_8H_{18}$ molecules is denoted $N_A$,
* The number of $O_2$ molecules is denoted $N_B$. 
* The number of $CO_2$ molecules is denoted $N_C$
* The number of $H_2O$ is denoted $N_D$.

The reaction rate is gonna be proportional to the collision frequency. Kinetic gas theory tells us that the collision frequency between the reactants is

$$
f_c = \sigma \bar{v}_{rel} \frac{N_A N_B}{V}
$$

where $\sigma$ is the **collision cross cross section** and $\bar{v}_{rel}$ is the **mean relative speed**. The mean relatve speed is given by

$$
\bar{v}_{rel} = \left\( \frac{8 k T}{\pi \mu} \right\)^{1/2}
$$

where

$$
\mu = \frac{m_A m_B}{m_A + m_B}.
$$

where $m_A$ and $m_B$ are the molecular masses of the two reactants respectively.

The reaction rate will also be proportional to the fraction of molecules having sufficient energy to react. This fraction can be calculated from the Boltzman distribution to be or the order

$$
e^{-E_{min}/kT}
$$

Putting it together we have the rate constant

$$
c = \sigma \bar{v}_{vel} \frac{N_A N_B}{V} e^{-E_{min}/k_BT}
$$


The code below should not be very hard to understand. If it is, it is probably wrong.

There is a posibility to vary the volume during the combustion. However, I have not done this 
yet because I find it unwise to do that before I am sure about the simpler part of the model.

```python
kB = 1.3807e-23     # Boltzmanns constant
d = 3e-10           # average molecule diameter
u = 1.660538921e-27 # Atomic mass unit
mA = (8*12+18)*u    # C8H18 molecular weight
mB = 2*16*u         # O2 molecular weight
mC = (12+2*16)*u    # CO2 molecular weight
mD = (2+16)*u       # H2O molecular weight
E = 46.8e6          # energy content (J/kg) of gasoline
M = 1e-3            # mass of petrol to simulate (kg)
V = 50e-6           # box volume
lam = 1.0           # Air-fuel-ratio to Stochiometric ratio ratio
Emin = 2e-19        # Activation energy

NA0 = M/mA          # initial number of C8H18
NB0 = lam*NA0*25./2 # initial number of O2
NC0 = 0.            # initial number of CO2
ND0 = 0.            # initial number of H2O

k2 = E*2./16*mA     # energy released for every CO2 created
mu = mA*mB/(mA+mB)

from scipy.stats import gamma

def dNdt(N,t=0):
    """ The differential equation governing the system """
    U = k2*N[2]     # internal energy (directly related to temperature) 
    T = 2./3.*U/kB/sum(N)+293.0
    vrel = (8*kB*T/mu/pi)**0.5
    p = 1.0-gamma.cdf(Emin,3./2,scale=kB*T)  # Fraction of molecules having enough energy (Boltzman)
    c = d*vrel/V*p
    #print c
    return array([-2*c*N[0]*N[1],
                  -25*c*N[0]*N[1],
                  16*c*N[0]*N[1],
                  18*c*N[0]*N[1]])

dNdt([NA0,NB0,NC0,ND0])
```

```python
from scipy import integrate
t = linspace(0, 1e-6,  1000)
N0 = array([NA0,NB0,NC0,ND0])
N, infodict = integrate.odeint(dNdt, N0, t, full_output=True)
infodict['message']
```

```python
NA, NB, NC, ND = N.T
plot(t, NA, 'r-', label='Gasolene')
plot(t, NB, 'b-', label='Oxygen')
plot(t, NC, 'g-', label='Carbon Dioxide')
plot(t, ND, 'y-', label='Water')
grid()
legend(loc='best')
xlabel('time')
ylabel('population')
title('Evolution of species');
```

```python
U = k2*NC
T = 2./3.*U/kB/(NA+NB+NC+ND)+293.0
plot(t,T,label='Temperature')
xlabel('time (s)')
ylabel('temperature (K)');
```

```python
Ntot = NA+NB+NC+ND
p = Ntot/V*kB*T
plot(t,p/100000.,label='Pressure')
xlabel('time (s)')
ylabel('pressure (Bar)');
```

# Boltzman

Gamma distribution pdf:
$$
p(x) = \frac{1}{\Gamma(k) \theta^k} x^{k - 1} e^{-\frac{x}{\theta}}
$$

Boltzman energy distribution:
p(E) = \frac{1}{\Gamma(3/2) (kT)^{3/2}} E^{1/2} e^{-\frac{E}{kT}}

Hence the Gamma parameters are

$$
x = E
$$

$$
\theta = k_BT
$$

$$
k=3/2
$$

```python
from scipy.stats import gamma
kB = 1.3807e-23 # Boltzmanns constant
T = 300.
Emin = linspace(0,2.5e-20,100)
T1 = 300.
p1 = lambda E: 1.0-gamma.cdf(E,3./2,scale=kB*T1)
T2 = 500.
p2 = lambda E: 1.0-gamma.cdf(E,3./2,scale=kB*T2)
plot(Emin,p1(Emin),label='%d K, Boltzman'%T1)
plot(Emin,p2(Emin),label='%d K, Boltzman'%T2)
plot(Emin,exp(-Emin/kB/T1),label='%d K, exp(-Emin/kT)' % T1)
plot(Emin,exp(-Emin/kB/T2),label='%d K, exp(-Emin/kT)' % T2)
xlabel('Required energy $E_{min}$ (J)')
ylabel('Fraction of molecules having required energy')
legend();
```
