## Experiments with uncertainty in S-N diagrams

Uncertainty in S-N diagrams follows a lognormal distribution. According to the literature the standard deviation is on the order of 0.2. This purpose of this notebook is to investigate what this means in practice.

First, lets see if we can generate som lognormally distributed numbers and take a look at the distribution. 

```python
from pylab import *

st = 0.2
mt = 400e3
x = lognormal(mean=log(10)*log10(mt),sigma=log(10)*st,size=10000)
hist(x,bins=100)

print "Check that we can reproduce our input:"
print "     Mean:", 10**mean(log10(x))
print "     Std:", std(log10(x))
```

That worked well. 

Now, lets look how it looks on an S-N curve

```python
from dafatigue import SN

N = logspace(0,7,100)

SU=300.0
SF=70.0
SU1=290.0
N1=130.
N2=2000000.0

figure(figsize=(9,6))
loglog(N,SN(N, SU=SU, SF=SF, SU1=SU1, N1=N1, N2=N2, va=False))
x = lognormal(mean=log(10)*log10(mt),sigma=log(10)*st,size=7)
S = SN(mt, SU=SU, SF=SF, SU1=SU1, N1=N1, N2=N2, va=False)
plot(x,S*ones(x.shape),'.')

from scipy.stats import lognorm
x = linspace(1.6e5,9e5,1000)
rv = lognorm(0.2,scale=400e3)
plot(x,S+5e6*rv.pdf(x))
axis([1e0,1e7,60,400])
```
